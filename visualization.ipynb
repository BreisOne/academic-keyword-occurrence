```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file in a DataFrame
data = pd.read_csv('out.csv')

# Sort the DataFrame by the "resultados" column in descending order
data = data.sort_values(by='results', ascending=True)

# Extract columns: keyword and results
keywords = data['search_term']
results = data['results']
```

```python
# Create a horizontal bar plot
plt.figure(figsize=(10, 12))  # Size of graph
plt.barh(keywords, results, color='skyblue')
plt.xlabel('Number of results')
plt.title('Results by keywords in google academic')

# Save bar plot in a pdf file
plt.savefig('bar_plot_results.pdf', format='pdf', bbox_inches ="tight")

# Show bar plot
plt.show()
```
